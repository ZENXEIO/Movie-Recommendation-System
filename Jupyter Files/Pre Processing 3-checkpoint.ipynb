```python
import pandas as pd
import numpy as np
from tmdbv3api import TMDb
import json
import requests
tmdb = TMDb()
tmdb.api_key = '94fc0930e034532d61661f5f8efe1bf8'
from tmdbv3api import Movie
tmdb_movie = Movie()
```

```python
def given_out_put_link(link_of_website):
    Main_df1 = pd.read_html(link_of_website, header=0)[2]
    Main_df2 = pd.read_html(link_of_website, header=0)[3]
    Main_df3 = pd.read_html(link_of_website, header=0)[4]
    Main_df4 = pd.read_html(link_of_website, header=0)[5]
    Main_df = pd.concat([Main_df1, Main_df2, Main_df3, Main_df4], axis=0)
    
    
    def get_genre2(x):
        genres = []
        result = tmdb_movie.search(x)
        movie_id = result[0].id
        response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}'.format(movie_id,tmdb.api_key))
        data_json = response.json()
        if data_json['genres']:
            genre_str = " " 
            for i in range(0,len(data_json['genres'])):
                genres.append(data_json['genres'][i]['name'])
                return genre_str.join(genres)
            else:
                np.NaN
        
        
    Main_df['genres'] = Main_df['Title'].map(lambda x: get_genre2(str(x)))
        
    Main_Dataset_of_given_date = Main_df[['Title', 'Cast and crew', 'genres']]
        
    def get_directors(x2):
        if " (director)" in x2:
            return x2.split(" (director)")[0]
        elif " (directors)" in x2:
            return x2.split(" (directors)")[0]
        else:
            return x2.split(" (director/screenplay)")[0]
        
    def get_act1(x3):
        return((x3.split("screenplay); ")[-1]).split(", ")[0])
    
    def get_act2(x4):
        if len((x4.split("screenplay); ")[-1]).split(", ")) < 2:
            return np.NaN
        else:
            return((x4.split("screenplay); ")[-1]).split(", ")[1])
    
    def get_act3(x5):
        if len((x5.split("screenplay); ")[-1]).split(", ")) < 3:
            return np.NaN
        else:
            return((x5.split("screenplay); ")[-1]).split(", ")[2])
            
    Main_Dataset_of_given_date['director_name'] = Main_Dataset_of_given_date['Cast and crew'].map(lambda x2: get_directors(x2))
    Main_Dataset_of_given_date['actor_1_name'] = Main_Dataset_of_given_date['Cast and crew'].map(lambda x3: get_act1(x3))
    Main_Dataset_of_given_date['actor_2_name'] = Main_Dataset_of_given_date['Cast and crew'].map(lambda x4: get_act2(x4))
    Main_Dataset_of_given_date['actor_3_name'] = Main_Dataset_of_given_date['Cast and crew'].map(lambda x5: get_act3(x5))
    
    Main_Dataset_of_given_date = Main_Dataset_of_given_date.rename(columns={'Title':'movie_title'})
    
    New_Data = Main_Dataset_of_given_date.loc[:,['director_name', 'actor_1_name', 'actor_2_name', 'actor_3_name', 'genres', 'movie_title']]
    
    
    New_Data["director_name"].fillna("Unknown", inplace = True)
    New_Data["actor_1_name"].fillna("Unknown", inplace = True)
    New_Data["actor_2_name"].fillna("Unknown", inplace = True)
    New_Data["actor_3_name"].fillna("Unknown", inplace = True)
    
    New_Data['movie_title'] = New_Data['movie_title'].str.lower()
    
    New_Data['Tags'] = New_Data['actor_1_name'] + ' ' + New_Data['actor_2_name'] + ' ' + New_Data['actor_3_name']+ ' ' + New_Data['director_name'] + ' ' + New_Data['genres']
    
    New_Data.drop_duplicates(subset ='movie_title', keep = 'last', inplace=True)
    
    return New_Data

    
```

```python
def given_out_put_link2(link_of_website):
    source = urllib.request.urlopen(link_of_website).read()
    soup = bs.BeautifulSoup(source,'lxml')
    
    tables = soup.find_all('table',class_='wikitable sortable')
    
    Main_df1 = pd.read_html(str(tables[0]))[0]
    Main_df2 = pd.read_html(str(tables[1]))[0]
    Main_df3 = pd.read_html(str(tables[2]))[0]
    Main_df4 =pd.read_html(str(tables[3]).replace("'1\"\'",'"1"'))[0]
    Main_df = pd.concat([Main_df1, Main_df2, Main_df3, Main_df4], axis=0)
    
    
    
    
    def get_genre2(x):
        genres = []
        result = tmdb_movie.search(x)
        if not result:
            return np.NaN
        else:
            movie_id = result[0].id
            response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}'.format(movie_id,tmdb.api_key))
            data_json = response.json()
            if data_json['genres']:
                genre_str = " " 
                for i in range(0,len(data_json['genres'])):
                    genres.append(data_json['genres'][i]['name'])
                    return genre_str.join(genres)
                else:
                    return np.NaN
        
        
    Main_df['genres'] = Main_df['Title'].map(lambda x: get_genre2(str(x)))
        
    Main_Dataset_of_given_date = Main_df[['Title', 'Cast and crew', 'genres']]
        
    def get_directors(x2):
        if " (director)" in x2:
            return x2.split(" (director)")[0]
        elif " (directors)" in x2:
            return x2.split(" (directors)")[0]
        else:
            return x2.split(" (director/screenplay)")[0]
        
    def get_act1(x3):
        return((x3.split("screenplay); ")[-1]).split(", ")[0])
    
    def get_act2(x4):
        if len((x4.split("screenplay); ")[-1]).split(", ")) < 2:
            return np.NaN
        else:
            return((x4.split("screenplay); ")[-1]).split(", ")[1])
    
    def get_act3(x5):
        if len((x5.split("screenplay); ")[-1]).split(", ")) < 3:
            return np.NaN
        else:
            return((x5.split("screenplay); ")[-1]).split(", ")[2])
            
    Main_Dataset_of_given_date['director_name'] = Main_Dataset_of_given_date['Cast and crew'].map(lambda x2: get_directors(x2))
    Main_Dataset_of_given_date['actor_1_name'] = Main_Dataset_of_given_date['Cast and crew'].map(lambda x3: get_act1(x3))
    Main_Dataset_of_given_date['actor_2_name'] = Main_Dataset_of_given_date['Cast and crew'].map(lambda x4: get_act2(x4))
    Main_Dataset_of_given_date['actor_3_name'] = Main_Dataset_of_given_date['Cast and crew'].map(lambda x5: get_act3(x5))
    
    Main_Dataset_of_given_date = Main_Dataset_of_given_date.rename(columns={'Title':'movie_title'})
    
    New_Data = Main_Dataset_of_given_date.loc[:,['director_name', 'actor_1_name', 'actor_2_name', 'actor_3_name', 'genres', 'movie_title']]
    
    
    New_Data["director_name"].fillna("Unknown", inplace = True)
    New_Data["actor_1_name"].fillna("Unknown", inplace = True)
    New_Data["actor_2_name"].fillna("Unknown", inplace = True)
    New_Data["actor_3_name"].fillna("Unknown", inplace = True)
    
    New_Data['movie_title'] = New_Data['movie_title'].str.lower()
    
    New_Data['Tags'] = New_Data['actor_1_name'] + ' ' + New_Data['actor_2_name'] + ' ' + New_Data['actor_3_name']+ ' ' + New_Data['director_name'] + ' ' + New_Data['genres']
    
    New_Data.drop_duplicates(subset ='movie_title', keep = 'last', inplace=True)
    
    return New_Data
```
